# 第4次作业 20201101 利用 Pyecharts 将链家房源信息制作成相关可视化图表

```python
# 终于把pyecharts装起来了，折腾了两天
import pandas as pd
import numpy as np
import pyecharts
from pyecharts import options as opts
from pyecharts.charts import Bar,Line,Pie
```

#### TODO1 ：
读取 Task_4_lianjia_sale.csv 数据(阅读参考文章 2、3)，利用 groupby 函数，以【区域】字段进行 groupby， 同时结合 agg 聚合函数统计各个区域平均房价;

将统计所得平均房价命名为【区域平均价格】列 (阅读参考文章 4、5)

并将数据从小到大按顺序排列(提示:df.sort_values(by = “ ”) (阅读参考文章 6));

最后展示所得前 5 行数据 (将所得的 DataFrame 命名为 df_avg);

```python
df = pd.read_csv('Task_4_lianjia_sale.csv ')
df_avg = df.groupby('区域')['价格'].agg([('区域平均价格','mean')]).sort_values(by ='区域平均价格' )
df_avg
```

#### TODO2:
利用 Pyecharts 将第(1)问所得数据作条形图(阅读参考文章 7)，图表标题为“上海各区域平均价格”(提示: 最后用 render_notebook()渲染显示出图片)

```python
x_1 = df_avg.index.values.tolist()
y_1 = df_avg["区域平均价格"].values.tolist()
y_1_ave = np.round(y_1).tolist()

bar = Bar()
bar.add_xaxis(x_1)
bar.add_yaxis("区域平均价格", y_1_ave)
bar.set_global_opts(title_opts=opts.TitleOpts(title="上海各区域平均价格"))
bar.render_notebook()
```

#### TODO3:
利用 groupby 函数，以【装修类型】字段进行 groupby，同时结合 agg 聚合函数统计各个区域房源数量;

将统计所得数据命名为【房源数量】列;最后展示所得数据(将所得的 DataFrame 命名为 df_zhuangxiu);

```python
df_zhuangxiu = df.groupby('装修类型')['区域'].agg([('房源数量','count')])
df_zhuangxiu
```

#### TODO4:
利用 Pyecharts 将第(3)问所得数据作饼图(阅读参考文章 7、8)，图表标题为“装修类型占比情况”(提示: v1=["啤酒","可乐","雪碧","咖啡","奶茶"] v2=[30,19,21,12,18]，a = [list(z) for z in zip(v1,v2)]， 则生成 v1 与 v2 一一对应的数据)。

```python
x_2 = df_zhuangxiu.index.values.tolist()
y_2 = df_zhuangxiu["房源数量"].values.tolist()

pie = Pie()
pie.add("", [list(z) for z in zip(x_2, y_2)])
pie.set_global_opts(title_opts=opts.TitleOpts(title="装修类型占比情况"))
pie.set_series_opts(label_opts=opts.LabelOpts(formatter="{b}:{d}%"))
pie.render_notebook()
```
