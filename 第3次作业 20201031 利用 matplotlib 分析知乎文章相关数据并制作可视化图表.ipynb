#### TODO1:
  读取 CSV 格式文件“task_3.1_zhihu_timeline_answer.csv”，将数据命名为 zhihu_data_answer，并查看前部数据(阅读参考文章 1、2)

```python
import pandas as pd
zhihu_data_answer = pd.read_csv('task_3.1_zhihu_timeline_answer.csv')
zhihu_data_answer.head()
```

#### TODO2:
  提取“task_3.1_zhihu_timeline_answer.csv”中【回答点赞数】和【回答感谢数】列数据，制成散点图， 观察两者的关系(阅读参考文章 3、4)

```python
import matplotlib.pyplot as plt
x0 = zhihu_data_answer['回答点赞数'].values.tolist()
y0 = zhihu_data_answer['回答感谢数'].values.tolist()
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False
plt.rcParams['savefig.dpi'] = 1000
plt.figure(figsize = (9, 6)) 
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel('问题回答数', fontsize=12)
plt.ylabel('问题关注数',fontsize=12)
plt.title('制作散点图',fontsize=20) 

plt.scatter(x0,y0,s=60,c='c',alpha = 0.25)

plt.show()
```

#### TODO3:
提取“task_3.1_zhihu_timeline_answer.csv”中【问题回答数】和【问题关注数】列数据，制成散点图， 观察两者的关系

```python
x1 = zhihu_data_answer['问题回答数'].values.tolist()
y1 = zhihu_data_answer['问题关注数'].values.tolist()
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False
plt.rcParams['savefig.dpi'] = 1000
plt.figure(figsize = (9, 6)) 
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel('问题回答数', fontsize=15)
plt.ylabel('问题关注数',fontsize=15)
plt.title('制作散点图',fontsize=20) 

plt.scatter(x1,y1,s=60,c='g',alpha = 0.25)
plt.show()
```

#### TODO4:

```python
zhihu_data_article = pd.read_csv('task_3.2_zhihu_timeline_article.csv')
zhihu_data_article.head()
```

#### TODO5:
提取“task_3.2_zhihu_timeline_article.csv”中【文章点赞数】和【文章评论数】列数据，制成散点图， 观察两者的关系

```python
x2 = zhihu_data_article['文章点赞数'].values.tolist()
y2 = zhihu_data_article['文章评论数'].values.tolist()
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False
plt.rcParams['savefig.dpi'] = 100000
plt.figure(figsize = (9, 6)) 
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel('文章点赞数', fontsize=15)
plt.ylabel('文章评论数',fontsize=15)
plt.title('制作散点图',fontsize=20) 
plt.scatter(x2,y2,s=40,c='c',alpha = 0.3)
plt.show()
```

#### TODO6:
提取“task_3.2_zhihu_timeline_article.csv”中【文章点赞数】低于 1000 的行数据，作出箱型图(阅读参考文章 5)

```python
x_4 = zhihu_data_article['文章点赞数'].values.tolist()
x4 = []
for i in x_4:
    if i < 1000:
        x4.append(i)
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False
plt.rcParams['savefig.dpi'] = 1000
plt.figure(figsize = (9, 6)) 
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

plt.boxplot(x4, notch=None, sym=None, vert=None,   
             whis=None, positions=None, widths=None,   
             patch_artist=None, meanline=None, showmeans=None,   
             showcaps=None, showbox=None, showfliers=None,   
             boxprops=None, labels=None, flierprops=None,   
             medianprops=None, meanprops=None,   
             capprops=None, whiskerprops=None)  
plt.ylabel('文章点赞数',fontsize=15)
plt.title('箱线图',fontsize=20.0)
plt.show()
```
